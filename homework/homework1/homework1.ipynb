```python
from scipy import stats

x = dataFrame["TV_advertisement"]
y = dataFrame["Sales"]
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
```

```python
# If you put 200 unit money on TV_advertisement you probably get 18 sales in return.
regression_line = slope * 200 + intercept
regression_line
```
