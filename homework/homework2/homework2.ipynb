```python
# Gerekli kütüphanelerin yüklenmesi
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np

# Verilen veri noktaları
points = np.array([[1, 2], [1, 4], [1, 0], [10, 2], [10, 4], [10, 0]])

# 1. Adım: Veri noktalarının görselleştirilmesi
plt.scatter(points[:, 0], points[:, 1], s=100, c='blue', label='Data Points')
plt.title("Verilen Veri Noktaları")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python

# 2. Adım: K-Means ile küme sayısını belirlemek için Elbow yöntemini kullanma
distortions = []
K = range(1, 6)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42).fit(points)
    distortions.append(kmeans.inertia_)
```

```python
# Elbow yöntemi grafiği
plt.plot(K, distortions, 'bx-')
plt.title("Elbow Yöntemi")
plt.xlabel("Küme Sayısı")
plt.ylabel("Distortion (İnertia)")
plt.show()
```

```python
# Elbow yöntemine göre küme sayısı: 2 seçiliyor
n_clusters = 2
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
kmeans.fit(points)
cluster_labels = kmeans.labels_
```

```python
# 3. Adım: Küme sonuçlarının görselleştirilmesi
plt.scatter(points[:, 0], points[:, 1], c=cluster_labels, s=100, cmap='viridis', label='Clusters')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label='Centers', marker='X')
plt.title("Kümeleme Sonuçları")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
# 4. Adım: Yeni veri noktalarının tahmini
new_points = np.array([[2, 2], [13, 5]])
new_labels = kmeans.predict(new_points)
```

```python
# Yeni veri noktalarını görselleştirerek kümelerde gösterme
plt.scatter(points[:, 0], points[:, 1], c=cluster_labels, s=100, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label='Centers', marker='X')
plt.scatter(new_points[:, 0], new_points[:, 1], s=200, c='black', label='New Points', marker='*')
plt.title("Yeni Noktalar ve Kümeler")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
# Tahmin edilen kümeler:
print(f"{new_points[0]} noktası {new_labels[0]} kümesine atanmıştır.")
print(f"{new_points[1]} noktası {new_labels[1]} kümesine atanmıştır.")
```
